**8. Operadores Aritméticos**

(Operador) | Descrição
* (+) | Adição
* (-) | Subtração
* (*) | Multiplicação
* (/) | Divisão
* (//) | Divisão inteira
* (%) | Resto
* (**) | Potência

```python
a = 10
b = 3

print(a + b)
print(a * b)
print(a ** b)
```
