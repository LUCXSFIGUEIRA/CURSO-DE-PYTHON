## 1) O básico ainda pode ser útil

Mesmo coisas simples como variáveis, listas, laços e funções já resolvem muitos problemas do dia a dia.


```
# Variáveis e operações simples
nome = "Lucas"
idade = 24
altura = 1.78

print(f"Nome: {nome}")
print(f"Idade: {idade}")
print(f"Altura: {altura} m")
print(f"Ano de nascimento aproximado: {2026 - idade}")
```

```
# Listas, laços e funções
numeros = [5, 12, 8, 21, 3, 18]

def resumo(lista):
    return {
        "quantidade": len(lista),
        "maior": max(lista),
        "menor": min(lista),
        "soma": sum(lista),
        "média": sum(lista) / len(lista)
    }

resultado = resumo(numeros)
resultado
```

## 2) Manipulação de texto

Python é muito forte para trabalhar com texto:
- limpar dados
- validar padrões
- procurar palavras
- extrair informações
- gerar mensagens automáticas


```
texto = "Python é uma linguagem poderosa, versátil e muito usada em automação."

print(texto.upper())
print(texto.lower())
print(texto.replace("automação", "análise de dados"))
print("versátil" in texto)
```

```
# Exemplo útil: contar palavras
frase = "python python dados automação python excel"
palavras = frase.split()

contagem = {}
for palavra in palavras:
    contagem[palavra] = contagem.get(palavra, 0) + 1

contagem
```

```
# Expressões regulares: encontrar padrões em texto
import re

mensagem = "Pedido 4589 entregue em 27/03/2026 para o cliente Lucas."
numero_pedido = re.findall(r"\d{4}", mensagem)
data = re.findall(r"\d{2}/\d{2}/\d{4}", mensagem)

print("Pedido encontrado:", numero_pedido)
print("Data encontrada:", data)
```

## 3) Automação de tarefas repetitivas

Uma das coisas mais legais do Python é automatizar trabalho manual.

Exemplos:
- renomear arquivos
- organizar pastas
- gerar relatórios
- tratar planilhas
- montar mensagens


```
# Exemplo: renomeando nomes de arquivos de forma simulada
arquivos = ["foto1.jpg", "foto2.jpg", "foto3.jpg"]

novos_nomes = []
for i, arquivo in enumerate(arquivos, start=1):
    extensao = arquivo.split(".")[-1]
    novos_nomes.append(f"evento_2026_{i:03d}.{extensao}")

novos_nomes
```

```
# Exemplo: gerar mensagens personalizadas automaticamente
clientes = ["Ana", "Bruno", "Carla"]

for cliente in clientes:
    print(f"Olá, {cliente}! Seu pedido já está disponível para retirada.")
```

## 4) Trabalhando com arquivos

Python consegue criar, ler e editar arquivos com facilidade.
Isso é muito usado para logs, relatórios, exportações e integrações.


```
# Escrevendo um arquivo de texto
conteudo = """Relatório de vendas
- Produto A: 10 unidades
- Produto B: 6 unidades
- Produto C: 14 unidades
"""

with open("relatorio.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write(conteudo)

print("Arquivo criado com sucesso!")
```

```
# Lendo o arquivo criado
with open("relatorio.txt", "r", encoding="utf-8") as arquivo:
    texto_lido = arquivo.read()

print(texto_lido)
```

## 5) Análise de dados com pandas

Aqui começa uma das áreas mais populares do Python.

Com `pandas`, você pode:
- ler planilhas e CSVs
- filtrar dados
- calcular indicadores
- resumir resultados
- preparar relatórios


```
import pandas as pd

dados = pd.DataFrame({
    "produto": ["Camiseta", "Calça", "Tênis", "Boné", "Camiseta"],
    "categoria": ["Roupas", "Roupas", "Calçados", "Acessórios", "Roupas"],
    "valor": [59.90, 129.90, 249.90, 39.90, 69.90],
    "quantidade": [3, 2, 1, 4, 5]
})

dados
```

```
# Criando uma coluna calculada
dados["faturamento"] = dados["valor"] * dados["quantidade"]
dados
```

```
# Resumindo informações
resumo_categoria = dados.groupby("categoria")[["quantidade", "faturamento"]].sum()
resumo_categoria
```

```
# Filtrando dados
dados[dados["faturamento"] > 200]
```

## 6) Visualização de dados

Com Python dá para criar gráficos para enxergar melhor os dados.


```
import matplotlib.pyplot as plt

resumo = dados.groupby("produto")["faturamento"].sum().sort_values()

plt.figure(figsize=(8, 4))
plt.barh(resumo.index, resumo.values)
plt.title("Faturamento por produto")
plt.xlabel("Faturamento")
plt.ylabel("Produto")
plt.show()
```

## 7) Simulações e cálculos

Python é ótimo para matemática, estatística, probabilidade e simulações.


```
# Simulação simples: lançamento de dado
import random
from collections import Counter

resultados = [random.randint(1, 6) for _ in range(1000)]
frequencias = Counter(resultados)

frequencias
```

```
plt.figure(figsize=(7, 4))
plt.bar(frequencias.keys(), frequencias.values())
plt.title("Frequência dos resultados em 1000 lançamentos")
plt.xlabel("Face do dado")
plt.ylabel("Quantidade")
plt.show()
```

```
# Exemplo prático: cálculo de IMC
def calcular_imc(peso, altura):
    return peso / (altura ** 2)

peso = 115
altura = 1.78
imc = calcular_imc(peso, altura)

print(f"IMC: {imc:.2f}")
```

## 8) Programação orientada a objetos

Quando o projeto cresce, você pode organizar melhor o código usando classes.


```
class Produto:
    def __init__(self, nome, preco, estoque):
        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def vender(self, quantidade):
        if quantidade <= self.estoque:
            self.estoque -= quantidade
            return f"Venda realizada: {quantidade} unidade(s) de {self.nome}"
        return "Estoque insuficiente"

    def resumo(self):
        return {
            "nome": self.nome,
            "preço": self.preco,
            "estoque": self.estoque
        }

produto = Produto("Camiseta", 59.90, 10)
print(produto.vender(3))
produto.resumo()
```

## 9) Pequeno projeto: ranking de vendas

Agora um exemplo mais próximo de algo real: descobrir os produtos mais vendidos.


```
vendas = pd.DataFrame({
    "produto": ["Camiseta", "Calça", "Tênis", "Boné", "Camiseta", "Tênis", "Boné", "Boné"],
    "quantidade": [3, 2, 1, 4, 5, 2, 1, 3]
})

ranking = (
    vendas.groupby("produto")["quantidade"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

ranking
```

```
plt.figure(figsize=(7, 4))
plt.bar(ranking["produto"], ranking["quantidade"])
plt.title("Ranking de produtos por quantidade vendida")
plt.xlabel("Produto")
plt.ylabel("Quantidade vendida")
plt.show()
```

## 11) Criando lógica de decisão

Python também serve para montar regras de negócio.


```
def classificar_nota(nota):
    if nota >= 9:
        return "Excelente"
    elif nota >= 7:
        return "Bom"
    elif nota >= 5:
        return "Regular"
    else:
        return "Insuficiente"

for nota in [9.5, 7.2, 5.8, 3.9]:
    print(nota, "->", classificar_nota(nota))
```

## 12) Mini desafio: filtrando números pares

Exemplo curto, mas muito comum em lógica.


```
numeros = list(range(1, 21))
pares = [n for n in numeros if n % 2 == 0]
pares
```

## 13) Exemplo de função útil para o dia a dia

Uma função para calcular desconto.


```
def aplicar_desconto(valor, percentual):
    desconto = valor * (percentual / 100)
    total = valor - desconto
    return round(total, 2)

print(aplicar_desconto(250, 10))
print(aplicar_desconto(89.90, 15))
```
